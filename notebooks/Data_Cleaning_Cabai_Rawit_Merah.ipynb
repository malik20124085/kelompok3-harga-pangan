```python
import pandas as pd

df = pd.read_excel("Downloads/Tabel Harga Berdasarkan Daerah.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
(df == "-").sum().sum()
```

```python
import numpy as np

df.replace("-", np.nan, inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
for col in df.columns[2:]:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
    )

    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
```

```python
df.describe()
```

```python
df_long = df.melt(
    id_vars=["No", "Komoditas (Rp)"],
    var_name="Tanggal",
    value_name="Harga"
)

df_long.head()
```

```python
df_long.shape
```

```python
df_long["Tanggal"] = pd.to_datetime(
    df_long["Tanggal"],
    dayfirst=True
)
```

```python
df_long.info()
```

```python
df_long = df_long.dropna(subset=["Harga"])
```

```python
df_long.isnull().sum()
```

```python
df_long.tail()
```

```python
print("Tanggal Awal :", df_long["Tanggal"].min())
print("Tanggal Akhir:", df_long["Tanggal"].max())
```

```python
df_long["Harga"].max()
```

```python
df_long["Harga"].min()
```

```python
df_long["Harga"].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df_long["Tanggal"], df_long["Harga"])
plt.title("Pergerakan Harga Cabai Rawit Merah")
plt.xlabel("Tanggal")
plt.ylabel("Harga")
plt.show()
```

```python
df_long.to_excel(
    "Data_Harga_Clean.xlsx",
    index=False
)
```
